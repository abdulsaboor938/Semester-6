```python
# 1- Reading the maze from a text file: Write a function to read the maze from a text file. The function
# should take a filename as input and return a 2D array representing the maze. The maze can be
# represented using a binary matrix, where 0 represents an empty cell and 1 represents a wall.

def read_maze(filename):
    maze = []
    with open(filename, 'r') as f:
        for line in f:
            maze.append([int(x) for x in line.split()])
    return maze

# 2- Implementing DFS and BFS: Write a function for DFS and BFS algorithms. Both functions should take
# the maze as input and return a path to the goal state if there exists one, and -1 if there is no path to
# the goal state.
#dfs 
def dfs(maze):
    stack = []
    stack.append((0,0))
    visited = set()
    visited.add((0,0))
    parent = {}
    parent[(0,0)] = None
    while stack:
        current = stack.pop()
        if current == (len(maze)-1, len(maze[0])-1):
            break
        for next in get_neighbors(current, maze):
            if next not in visited:
                stack.append(next)
                visited.add(next)
                parent[next] = current
    if current != (len(maze)-1, len(maze[0])-1):
        return -1
    path = []
    while current != None:
        path.append(current)
        current = parent[current]
    path.reverse()
    return path

#bfs
def bfs(maze):
    queue = []
    queue.append((0,0))
    visited = set()
    visited.add((0,0))
    parent = {}
    parent[(0,0)] = None
    while queue:
        current = queue.pop(0)
        if current == (len(maze)-1, len(maze[0])-1):
            break
        for next in get_neighbors(current, maze):
            if next not in visited:
                queue.append(next)
                visited.add(next)
                parent[next] = current
    if current != (len(maze)-1, len(maze[0])-1):
        return -1
    path = []
    while current != None:
        path.append(current)
        current = parent[current]
    path.reverse()
    return path

#get neighbors

def get_neighbors(current, maze):
    neighbors = []
    if current[0] > 0 and maze[current[0]-1][current[1]] == 0:
        neighbors.append((current[0]-1, current[1]))
    if current[0] < len(maze)-1 and maze[current[0]+1][current[1]] == 0:
        neighbors.append((current[0]+1, current[1]))
    if current[1] > 0 and maze[current[0]][current[1]-1] == 0:
        neighbors.append((current[0], current[1]-1))
    if current[1] < len(maze[0])-1 and maze[current[0]][current[1]+1] == 0:
        neighbors.append((current[0], current[1]+1))
    return neighbors

# 3- Returning the path to the goal state: The function should return a list of coordinates representing the
# path from the starting point to the goal state. If there is no path, the function should return -1.

def print_path(path):
    if path == -1:
        print("No path")
    else:
        print("Path: ", end="")
        for i in range(len(path)):
            if i == len(path)-1:
                print(path[i])
            else:
                print(path[i], end="->")

# 4- Bonus: Design the maze using Python graphics library: Students can create a graphic representation of
# the maze using Python graphics library. The graphic representation of the maze should include a start
# and goal point, and the path returned by the DFS or BFS algorithm. Students can use any Python
# graphics library of their choice.

def draw_maze(maze, path):
    import pygame
    pygame.init()
    width = 150
    height = 100
    margin = 5
    screen = pygame.display.set_mode([(width+margin)*len(maze[0])+margin, (height+margin)*len(maze)+margin])
    pygame.display.set_caption("Maze")
    done = False
    clock = pygame.time.Clock()
    while not done:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                done = True
        screen.fill((0,0,0))
        for i in range(len(maze)):
            for j in range(len(maze[0])):
                if maze[i][j] == 1:
                    pygame.draw.rect(screen, (255,255,255), [(margin+width)*j+margin, (margin+height)*i+margin, width, height])
        for i in range(len(path)):
            if i == 0:
                pygame.draw.rect(screen, (0,255,0), [(margin+width)*path[i][1]+margin, (margin+height)*path[i][0]+margin, width, height])
            elif i == len(path)-1:
                pygame.draw.rect(screen, (255,0,0), [(margin+width)*path[i][1]+margin, (margin+height)*path[i][0]+margin, width, height])
            else:
                pygame.draw.rect(screen, (0,0,255), [(margin+width)*path[i][1]+margin, (margin+height)*path[i][0]+margin, width, height])
        clock.tick(60)
        pygame.display.flip()
    pygame.quit()

# 5- Testing: Write a main function to test your code. The main function should read the maze from a text
# file, call the DFS and BFS functions, and print the path returned by the functions. The main function
# should also call the function to draw the maze using Python graphics library.
def main():
    maze = read_maze("maze.txt")
    print("DFS:")
    # path = dfs(maze)
    print_path(path)
    draw_maze(maze, path)
    print("BFS:")
    path = bfs(maze)
    print_path(path)
    draw_maze(maze, path)

if __name__ == "__main__":
    main()
```
