```python
# importing libraries

import pandas as pd

from sklearn.cluster import KMeans

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('driver.csv')
df.head()
```

```python
df.drop(['id'],axis=1,inplace=True)
```

```python
df.fillna(df.mean(),inplace=True)
df.head()
```

## Elbow Curve

```python
from sklearn.cluster import KMeans

wcss =[]

for i in range(1,11):
    algo = KMeans(n_clusters = i, n_init = 'auto')
    algo.fit(df)
    wcss.append(algo.inertia_)

# plotting the graph
import matplotlib.pyplot as plt
plt.plot(range(1,11),wcss)
plt.title('Elbow Curve')
plt.xlabel('Number of Clusters')
plt.ylabel('Within Sum of Squares')
plt.show()
```

## K-means Clustering

```python
from sklearn.cluster import KMeans
algo = KMeans(n_clusters = 4, n_init='auto')
algo.fit(df)
y_pred = algo.predict(df) # we can also use algo.labels_
```

```python
# plotting a simple graph with dfferent colors for different clusters
plt.scatter(df['mean_dist_day'],df['mean_over_speed_perc'],c=y_pred)
plt.show()
```

```python
# plot with cluster centers
plt.scatter(df['mean_dist_day'],df['mean_over_speed_perc'],c=algo.labels_)
plt.scatter(algo.cluster_centers_[:,0],algo.cluster_centers_[:,1],s=300,c='red',marker='x')
plt.show()
```
