```python
data = pd.read_excel('Q2_Clustering.xlsx', engine='openpyxl')
data.head()
```

```python
# Encoding categorical data
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()

for col in data.columns:
    data[col] = labelencoder.fit_transform(data[col])

data.head()
```

```python
# K-Means

from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
kmeans.fit(data)
labels = kmeans.predict(data)

labels
import matplotlib.pyplot as plt
plt.scatter(data.iloc[:,0], data.iloc[:,1], c=labels, cmap='rainbow')
plt.show()
```

```python
# Agglomerative Clustering

from sklearn.cluster import AgglomerativeClustering
agglomerative = AgglomerativeClustering(n_clusters=3, linkage='complete')
agglomerative.fit(data)

labels = agglomerative.labels_
labels

import matplotlib.pyplot as plt
plt.scatter(data.iloc[:,0], data.iloc[:,1], c=labels, cmap='rainbow')
plt.show()
```

```python
# dbscan

from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=2, min_samples=6)
dbscan.fit(data)

labels = dbscan.labels_
labels

import matplotlib.pyplot as plt
plt.scatter(data.iloc[:,0], data.iloc[:,1], c=labels, cmap='rainbow')
plt.show()
```

```python
# Birch

from sklearn.cluster import Birch
birch = Birch(n_clusters=3)
birch.fit(data)

labels = birch.predict(data)
labels

import matplotlib.pyplot as plt
plt.scatter(data.iloc[:,0], data.iloc[:,1], c=labels, cmap='rainbow')
plt.show()
```

# DBSCAN is worst because it does not differentitate and we can assume Agglomerative is the best because it is the only one that can differentiate between the two clusters, and is giving arbitratry shapes to the clusters.
