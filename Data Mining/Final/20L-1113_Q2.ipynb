# Preprocesing

```python
# importing data
import pandas as pd

# reading data
df = pd.read_excel('Q2_Clustering.xlsx')
df.head()
```

```python
# dropping unnecessary columns

# dropping CustomerID, Name, Address
df.drop(['CustomerID', 'Name', 'Address'], axis=1, inplace=True)
df.head()
```

```python
# Encoding categorical data

df['Gender'] = df['Gender'].astype('category')
df['Gender'] = df['Gender'].cat.codes

df['Occupation'] = df['Occupation'].astype('category')
df['Occupation'] = df['Occupation'].cat.codes

df['City'] = df['City'].astype('category')
df['City'] = df['City'].cat.codes

df.head()
```

```python
# Scaling data by min-max scaler

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
df_scaled.head()
```

## DBSCAN

```python
# db scan
from sklearn.cluster import DBSCAN
from sklearn.metrics import accuracy_score

model = DBSCAN(eps=0.3, min_samples=10)
model.fit(df)
pred = model.fit_predict(df)
pred

# Visualize the clustering results
import matplotlib.pyplot as plt

plt.scatter(df['Age'], df['Annual Income (k$)'], c=pred)
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('DBSCAN Clustering')
plt.show()
```
