```python
# exploratory data analysis
data = pd.read_excel('automobile_data.xlsx')

# first 5 rows of the data
data.head()
```

```python
data.columns
```

```python
# checking unique values in each column
data.nunique()
```

```python
# Check the missing values for each column.
data.isnull().sum()
```

```python
# Drop unnecessary data columns
new_data = data.drop(['engine-type','average-mileage'], axis=1)
new_data.head()
```

```python
# Drop null value
data.dropna(subset=data.columns,inplace=True)
data.head()
```

### 4.2.2

```python
# Find the most expensive car from the dataset and display its price
data[data['price'] == data['price'].max()]
```

```python
# Calculate total cars per manufacturer and show the result
data['company'].value_counts()
```

```python
# Read the details of vehicles against Toyota manufacturer and print them
data[data['company'] == 'toyota']
```

```python
# Arrange the cars according to the prices (highest-lowest) and display relevant information for first 5 rows only
data.sort_values(by='price', ascending=False).head()
```
